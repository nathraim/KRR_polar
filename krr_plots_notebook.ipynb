```python
def plot_polar_krrvsdfpt(file_path, file_name,amin,amax):
    import matplotlib.font_manager as font_manager

##################### Read Data #####################

    results_file_dir = os.path.join(file_path, file_name)
    results_file = open(results_file_dir,'r')
    my_file = 'plot_polar_krrvsdfpt.png'
        
        
    xx_actual = []
    yy_actual = []
    zz_actual = []
    xy_actual = []
    xz_actual = []
    yz_actual = []

    xx_predict = []
    yy_predict = []
    zz_predict = []
    xy_predict = []
    xz_predict = []
    yz_predict = []
    
    line_index = 0
    
    for line in results_file:
        
        line_index += 1
        
        if line_index == 1:
            error = list(map(float,line.split()[18:]))                  
        
        if line_index > 2:
            
                
            xx_actual.append(line.split()[1])
            xx_predict.append(line.split()[2])
            
            yy_actual.append(line.split()[3])
            yy_predict.append(line.split()[4])
            
            zz_actual.append(line.split()[5])
            zz_predict.append(line.split()[6])
        
            xy_actual.append(line.split()[7])
            xy_predict.append(line.split()[8])
            
            xz_actual.append(line.split()[9])
            xz_predict.append(line.split()[10])
            
            yz_actual.append(line.split()[11])
            yz_predict.append(line.split()[12])
            
      ##################### Plotting #####################
    
    plt.figure(figsize=(12,9))

    plt.plot(xx_actual, xx_predict, linestyle='',marker='o',markersize=9, color = 'r')
    plt.plot(yy_actual, yy_predict, linestyle='',marker='o',markersize=9, color = 'b')
    plt.plot(zz_actual, zz_predict, linestyle='',marker='o',markersize=9, color = 'lime')
    plt.plot(xy_actual, xy_predict, linestyle='',marker='o',markersize=9, color = 'gold')
    plt.plot(xz_actual, xz_predict, linestyle='',marker='o',markersize=9, color = 'm')
    plt.plot(yz_actual, yz_predict, linestyle='',marker='o',markersize=9, color = 'cyan')
    
    components = ['$xx$ ({:2.2f})'.format(error[0]),'$yy$ ({:2.2f})'.format(error[1]),'$zz$ ({:2.2f})'.format(error[2]),'$xy$ ({:2.2f})'.format(error[3]),'$xz$ ({:2.2f})'.format(error[4]),'$yz$ ({:2.2f})'.format(error[5])]

    plt.legend(components, bbox_to_anchor=(0.65, 0.55), loc='upper left',frameon=False,prop=font_manager.FontProperties(family='times new roman',weight='normal',style='normal', size=26))
    
    plt.xticks(fontsize=28,fontname='times new roman')
    plt.yticks(fontsize=28,fontname='times new roman')
    plt.xlim(amin,amax)
    plt.ylim(amin,amax)
    plt.xlabel('$\\alpha^{\mathrm{DFPT}}$', fontsize=32,fontname='times new roman')
    plt.ylabel('$\\alpha^{\mathrm{GPR}}$', fontsize=32,fontname='times new roman')
    
    x = [amin, amax]
    y = [amin, amax]
    
        
    plt.plot(x, y, '-', color = 'k')

    #plt.savefig(os.path.join(file_path+'/../plots', my_file))
    #plt.savefig('plots/'+'plot_polar_krrvsdfpt', bbox_inches ='tight', dpi=300)
    output = results_file_dir = os.path.join(file_path, my_file)
    plt.savefig(output, bbox_inches ='tight', dpi=300)

    plt.show()
```

```python
def plot_polar_series(file_path, file_name,xmin,xmax,ymin,ymax):
    import matplotlib.font_manager as font_manager
##################### Read Data #####################

    results_file_dir = os.path.join(file_path, file_name)
    results_file = open(results_file_dir,'r')
    
    number = []
    
    xx_actual = []
    yy_actual = []
    zz_actual = []
    xy_actual = []
    xz_actual = []
    yz_actual = []

    xx_predict = []
    yy_predict = []
    zz_predict = []
    xy_predict = []
    xz_predict = []
    yz_predict = []
    
    test=[]
    
    line_index = 0
    
    for line in results_file:
        
        line_index += 1
        
        if line_index > 2:
           
            test.append(line.split()[1:])
            number.append(line.split()[0])
            
            xx_actual.append(line.split()[1])
            xx_predict.append(line.split()[2])
            
            yy_actual.append(line.split()[3])
            yy_predict.append(line.split()[4])
            
            zz_actual.append(line.split()[5])
            zz_predict.append(line.split()[6])
        
            xy_actual.append(line.split()[7])
            xy_predict.append(line.split()[8])
            
            xz_actual.append(line.split()[9])
            xz_predict.append(line.split()[10])
            
            yz_actual.append(line.split()[11])
            yz_predict.append(line.split()[12])
            
      ##################### Plotting #####################
        
    test=np.array(test)
    
    for i_index in range(0,2*6,2):
        
        plt.figure(figsize=(12,9))

        plt.plot(number, test[:,i_index], '-', color = 'black',lw=2)
        plt.plot(number, test[:,i_index+1], '-', color = 'orange',lw=2)

        #plt.plot(number, zz_actual, '-', color = 'c')
        #plt.plot(number, zz_predict, '-', color = 'y')

        components = ['DFPT','GPR']

        plt.legend(components, bbox_to_anchor=(0.10, 0.97), loc='upper left', fontsize=24,frameon=False,prop=font_manager.FontProperties(family='times new roman',weight='normal',style='normal', size=26))

        plt.xlim(xmin,xmax)
        plt.ylim(ymin,ymax)
        plt.xticks(fontsize=28,fontname='times new roman')
        plt.yticks(fontsize=28,fontname='times new roman')
        plt.xlabel('time step', fontsize=32,fontname='times new roman')
        print(i_index)
        if (i_index==0):
            plt.ylabel('$\\alpha_{xx}$', fontsize=32,fontname='times new roman')
            plt.savefig('plot_polar_series_xx', bbox_inches ='tight', dpi=300)
        if (i_index==2):  plt.ylabel('$\\alpha_{yy}$', fontsize=32,fontname='times new roman')
        if (i_index==4):  plt.ylabel('$\\alpha_{zz}$', fontsize=32,fontname='times new roman')
        if (i_index==6):  plt.ylabel('$\\alpha_{xy}$', fontsize=32,fontname='times new roman')
        if (i_index==8):  plt.ylabel('$\\alpha_{xz}$', fontsize=32,fontname='times new roman')
        if (i_index==10):  plt.ylabel('$\\alpha_{yz}$', fontsize=32,fontname='times new roman')
        

        plt.show()
```

```python
def plot_raman(file_path, file_name, file_name2,xmin,xmax,ymax):

##################### Read Data #####################

    results_file_dir = os.path.join(file_path, file_name)
    results_file = open(results_file_dir,'r')
    my_file = 'plot_raman_krrvsdfpt.png'
    
        
    freqs = []
    intensity = []
      
    for line in results_file:
        
        freqs.append(line.split()[0])
        intensity.append(line.split()[1])
    
    results_file.close()
    
    results_file_dir = os.path.join(file_path, file_name2)
    results_file = open(results_file_dir,'r')
    
    freqs_krr = []
    intensity_krr = []
    
    for line in results_file:
        
        freqs_krr.append(line.split()[0])
        intensity_krr.append(line.split()[1])
        
    results_file.close()
            
      ##################### Plotting #####################
    
    plt.figure(figsize=(12,9))
    

    plt.plot(freqs, intensity, '-', lw=2.5, color = 'black')
    plt.plot(freqs_krr, intensity_krr, '-', lw=2.5, color = 'orange')
    plt.plot(markersize=5,markeredgewidth=5)
            
    components = ['DFPT','GPR']

    plt.legend(components, bbox_to_anchor=(0.55, 0.9), loc='upper left', fontsize=24, frameon=False)
    
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlim(xmin,xmax)
    plt.ylim(0,ymax)
    plt.xlabel('Wavenumber (cm$^{-1}$)', fontsize=28)
    plt.ylabel('$I(\\omega)$', fontsize=28)
    plt.gca().axes.get_yaxis().set_ticks([]) # Removes ticks for y axis
    
    #plt.savefig('plot_raman_krrvsdfpt', bbox_inches ='tight', dpi=300)
    #plt.savefig(os.path.join(file_path+'/../plots', my_file))
    output = results_file_dir = os.path.join(file_path, 'Raman.pdf')
    plt.savefig(output, bbox_inches ='tight', dpi=300)
    
    plt.show()
```
